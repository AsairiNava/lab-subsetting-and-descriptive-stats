#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
# your answer here
cities_between_15_and_20 = temp.loc[(temp['Temperature'] >= 15) & (temp['Temperature'] <= 20)]
cities_between_15_and_20 = cities_between_15_and_20.drop(['State'], axis=1)
cities_between_15_and_20
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
New_York_mean_temp = np.mean(temp.loc[temp['State'] == 'New York'])
print("New York temperature mean is", New_York_mean_temp)
New_York_std_temp = np.std(temp.loc[temp['State'] == 'New York'])
print("New York temperature standard deviation is", New_York_std_temp)

Connecticut_mean_temp = np.mean(temp.loc[temp['State'] == 'Connecticut'])
print("Connecticut temperature mean is", Connecticut_mean_temp)
Connecticut_std_temp = np.std(temp.loc[temp['State'] == 'Connecticut'])
print("Connecticut temperature standard deviation is", Connecticut_std_temp)

New_Jersey_mean_temp = np.mean(temp.loc[temp['State'] == 'New Jersey'])
print("Connecticut temperature mean is", New_Jersey_mean_temp)
New_Jersey_std_temp = np.std(temp.loc[temp['State'] == 'New Jersey'])
print("Connecticut temperature standard deviation is", New_Jersey_std_temp)
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# your answer here
max_dif = employee.groupby('Department').apply(lambda x: x[x['Salary'] == x['Salary'].max()])
min_dif = employee.groupby('Department').apply(lambda x: x[x['Salary'] == x['Salary'].min()])
print(max_dif, min_dif)
employee.groupby(['Department']).agg({'Salary':lambda x: x.max() - x.min()})
```
